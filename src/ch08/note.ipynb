### 예외 처리

#### 예외

* 예외(exception)란?
  * 프로그램 실행 도중에 오류가 발생하면 생성되는 특별한 객체
    * print(1/0)을 실행하면 ZeroDivisionError 예외 발생
  * 예외가 발생하면 파이썬은 프로그램 실행을 중단
  * 예외처리란?
    * 예외가 발생했을 때 프로그램 실행을 중단하지 않고 처리하는 방법을 정의한 프로그램 부분
  * 예외의 종류
    * ZeroDivisionError 0으로 나눌 때
    * IndexError        인덱스가 범위를 벗어날 때
    * FileNotFoundError 존재하지 않는 파일을 열려고 할 때
    * ValueError        원하는 값을 입력 받지 못할 때
    * NameError         정의되지 않은 변수를 사용할 때
    * TypeError         데이터형이 맞지 않는 연산을 할 때

* 여러가지 예외
  * ```Python
    >>> 5/0
    ZeroDivisionError: division by zero
    >>> a=[0,1,2]
    >>> a[3]
    IndexError: list index out of range
    >>> fp=open("Nofile.txt", "r")
    FileNotFoundError: [Errno 2] No such file or directory: 'Nofile.txt'
    >>> int(input('Type float number or string : '))
    Type float number or string : 1.23
    ValueError: invalid literal for int() with base 10: '1.23'
    >>> 1 + var*2
    NameError: name 'var' is not defined
    >>> "sr"+1
    TypeError: can only concatenate str (not "int") to str
    ```

#### 예외 처리

* 예외 처리(1)
  * try-except/try-except-else 문
    * ```Python
    try:
        a = 1 / b # 1번
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.") # 2번
    else:
        print(a) # 3번
    ```
  * 1번 a=1/b를 실행하고 ZeroDivisionError 예외가 발생하면 2번 문장 실행
  * ZeroDivisionError가 아닌 다른 예외가 발생하면 프로그램 중단
  * 예외가 발생하지 않으면 3번 문장 실행
  * else: 블록은 생략 가능
  * except ZeroDivisionError: 에서 ZeroDivisionError를 생략하면 모든 예외에 대해 2번 문장 실행
  
  * 예외를 알 수 없는 경우 try-except-else 문
    * ```Python
    try:
        a = int(input("Type a Number: ")) # 1번
    except Exception as e: # 2번
        print("예외가 발생했습니다", e) # 3번
    else:
        print(a) # 4번
    ```
  *  1번을 싱행하고 예외가 발생하면 3번 문장 실행
  *  예외의 종류는 모르지만 예외가 발생한 이유를 알고 싶을 때 2번과 같이 사용. e는 예외와 관련된 메시지 저장
  *  예외가 발생하지 않으면 4번 문장 실행
  
  * 복합 try-except 문
    * ```Python
    import sys
    try:
        fp = open('sample.txt')
        sl = fp.readline()
        value = int(sl.strip())
    except OSError as err:
        print("OS 오류: ", err) # 1번
    except ValueError:
        print("정수로 변환할 수 없습니다.") # 2번
    except:
        print("알 수 없는 오류가 발생하였습니다") # 3번
    ```
  * try: 블록 실행 결과 OSError 예외가 발생하면 1번 문장 실행
  * ValueError 예외가 발생하면 2번 문장 실행
  * 기타 예외가 발생하면 3번 문장 실행
  
* 예외 처리(2)
  * try-except-finally 문
    * ```Python
    try:
        result = x/ y # 1번
    except ZeroDivisionErrorL
        print("0으로 나눌 수 없습니다") # 2번
    except:
        print(result) # 3번
    finally:
        print("예외처리가 끝났습니다.") # 4번
     ```
  * 1번 문장을 싱핼하고 ZeroDivisionError 예외가 발생하면 2번 문장 실행
  * 기타 예외가 발생하면 3번 문장 실행
  * 예외 발생에 관계없이 4번 문장 실행. 따라서 4번 문장은 항상 실행됨
  * except-else 블록은 생략 가능(try-finally)

#### 사용자 예외 클래스

* 사용자 예외 클래스
  * 사용자가 예외 클래스를 선언하고, 함수에서 예외를 발생시키기
  * 사용자 예외 클래스는 Exception 클래스에서 상속받아 정의
  * __str__ 함수를 작성하면 예외 내용을 문자열로 출력 가능
  * __init__ 함수를 작성하면 예외를 발생시킬 때 값을 전달할 수 있음

```Python
class NegativeNumberError(Exception):
  def __init__(self, value):
    self.value = value
    self.message = f"음수는 허용되지 않습니다: 입력값({value})"
    super().__init__(self.message)

  def __str__(self):
    return self.message

def process_number(n):
  if n < 0:
    raise NegativeNumberError(n)
  return n * 2

try:
  num = int(input("양의 정수를 입력하세요: "))
  result = process_number(num)
except ValueError:
  print("숫자가 아닌 값을 입력했습니다.")
except NegativeNumberError as e:
  print(f"사용자 정의 예외 발생: {e}")
else:
    print(f"처리 결과: {result}")
finally:
    print("프로그램 종료.")
```


### 파이썬의 파일 처리

#### 파일 입출력

* 표준 입출력
  * 표준 입력 (standard input) : 키보드
  * 표준 출력 (standard output) : 콘솔 화면
  * 표준 에러 (standard error) : 콘솔 화면
* 파일 입출력
  * 표준 입출력 대신에 파일에 기록하는 것
  * 텍스트(text) 파일 : 사람이 볼 수 있는 문자 데이터를 저장 (메모장으로 확인 가능)
  * 이진(binary) 파일 : 이진수 데이터를 저장
* 파일 처리 과정
  * 열기 : 파일 이름과 처리 목적(읽기 또는 쓰기)를 지정해 사용하려는 파일을 열기
  * 사용 : 열린 파일에 대한 읽기 또는 쓰기를 진행
  * 닫기 : 파일의 사용이 끝나면 열린 파일을 닫아 버퍼에 있는 내용을 디스크에 모두 반영

#### 파일 다루기

* 파일의 개념적 모델
  * 오프셋 0부터 끝까지 1Byte 단위로 직렬화되어 저장됨
* 파일의 열기
  * fp = open("filename", "mode")
    * Filename : 열고자 하는 파일명 (절대경로 혹은 상대경로로 지정)
    * mode : r, w, a, x / r+, w+, a+ / rb, wb, ab, rb+, wb+, ab+

#### 파일 다루기 : 텍스트 파일

* 파일의 열기 모드 (텍스트 파일)
  
  |mode|설명|
  |----|----|
  |'r'|읽기 전용 모드. 파일이 존재해야 함 (파일 없으면 에러 발생).|
  |'w'|쓰기 전용 모드. 파일이 없으면 생성, 있으면 내용 덮어쓰기.|
  |'a'|추가 모드. 파일 끝에 내용 덧붙임. 파일이 없으면 파일 생성.|
  |'x'|베타적 생성 모드. 파일이 있으면 오류 발생, 없으면 새로 생성.|
  
  <br>

  |mode|설명|
  |----|----|
  |'r+'|읽기/쓰기 모드. 파일이 존재해야 함.|
  |'w+'|읽기/쓰기 모드. 파일 없으면 생성, 있으면 덮어쓰기.|
  |'a+'|읽기/쓰기 모드. 파일 끝에 덧붙이기. 기존 내용은 유지.|

#### 텍스트 파일 : 쓰기
* 현재 폴더(실행 중인 폴더)에 phonebook.txt 파일을 쓰기 모드로 열고 유니코드의 인코딩 방식은 utf-8 형식으로 저장하는 코드 예시
  * 코드 설명
    * UTF-8(Unicode Transfer Format 8-bit)
      * 가변 길이 인코딩 : 한 문자를 1~4바이트로 표현, ASCII 문자(영문, 숫자 등)는 1B, 한글/한자/특수문자 2~4B
      * 호환성 : ASCII와 완변 호환 (기존 시스템과 충돌 없음)
      * 범용성 : 유니코드 기반으로 거의 모든 언어의 문자를 표현 가능
      * 효율성 : 영어 중심의 문서에서는 저장 공간을 절약할 수 있음
    * write 함수 : 여러 줄을 쓸 때, 문자열로 출력(텍스트 파일에 출력)
    * writeline 함수 : 한 줄 씩 쓸 때
    * 한 줄 (line) : 제일 끝 문자가 '\n'으로 끝냄
    * 'w'모드 주의사항
      * 파일이 없으면 생성하고, 있으면 기존의 내용을 덮어쓰기하므로 주의

  ```Pyhton
  ofile = open("./phonebook.txt", "w", encoding="utf-8") # 열기

  ofile.write("John Doe, 123-456-7890\nJane Smith, 987-654-3210\n")
  ofile.writeline("Charlie Davis, 333-222-1111\n")
  ofile.writeline("Eve While, 777-888-9999\n")

  ofile.close() # 파일 닫기
  ```

#### 텍스트 파일 : 읽기

* 현재 폴더(실행 중인 폴더)에 유니코드의 인코딩 방식은 utf-8 형식으로 저장된 phonebook.txt 텍스트 파일을 읽기 모드로 열고 각 라인을 출력하는 코드 예시
  * 코드 설명
    * Utf-8(Unicode Transfer Format 8-bit) 형식을 디코딩하여 유니코드로 변환하여 읽음
    * readlines 함수 : 여러 줄을 읽어 리스트로 반환 (개행문자, '\n', new line 문자를 한 줄로 취급)
    * readline 함수 : 한 줄 씩 읽어 문자열로 반환
    * str.strip() : 앞 뒤의 공백 문자 제거
    * 만약 파일이 없으면 예외 발생

```Python
rfile = open("./phonebook.txt", "r", encoding="utf-8")
line : str = rfile.readline() # 한 줄 씩 읽어서 Line에 저장, Line : str는 타입 어노테이션을 이용하여 타입 힌트 제공
while line:
  print(line.strip()) # 앞 뒤 공백 문자 제거
  line = rfile.readline() # 한 줄 씩 읽어서 line에 저장
rfile.close

# or (단 파일 사이즈가 크면 메모리가 부족할 수 있어 앞의 것을 선호)
rfile = open("./phonebook.txt", "r", encoding="utf-8")
lines : list = rfile.readlines() # 여러 줄을 읽어서 리스트로 반환하여 Lines에 저장, Line : list는 타입 어노테이션을 이용하여 타입 힌트 제공
for line in lines:
  print(line.strip()) # 앞 뒤 공백 문자 제거
rfile.close
```
