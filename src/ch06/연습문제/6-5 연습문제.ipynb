* Articles 클래스를 작성하라. 이 클래스는 각각 제품의 이름, 재고 수량, 단가를 나타내는
name, stock, price 인스턴스 변수를 가지고 있다. offer_price() 메소드는 구매 수량을 인수
로 전달받아 총 가격을 반환한다. 구매 수량이 1~10개 이하이면 정상 가격을 반환하고, 11~50개
이면 5% 할인된 가격을 반환한다. 또 51개 이상이면 10% 할인해 준다. purchase() 메소드는
구매 수량을 인수로 전달받아 재고 수량을 구매 수량만큼 감소한다.

* 이 클래스를 이용하여 단가 1000원인 ”mask“를 10,000장으로 설정하고, 40장 구매 가격과
구매 후 재고를 출력하라.

```python
# -*- coding: utf-8 -*-
"""
연습문제 6-5: Articles 클래스 구현
"""

class Articles:
    """
    제품의 이름, 재고, 가격 정보를 관리하는 클래스.
    """
    def __init__(self, name: str, stock: int, price: int):
        """
        Articles 클래스의 인스턴스를 초기화합니다.

        Args:
            name (str): 제품의 이름.
            stock (int): 재고 수량.
            price (int): 단가.
        """
        self.name = name
        self.stock = stock
        self.price = price

    def offer_price(self, quantity: int) -> float:
        """
        구매 수량에 따라 할인된 총 가격을 계산하여 반환합니다.
        - 1~10개: 정상가
        - 11~50개: 5% 할인
        - 51개 이상: 10% 할인

        Args:
            quantity (int): 구매 수량.

        Returns:
            float: 할인율이 적용된 총 가격.
        """
        if not isinstance(quantity, int) or quantity <= 0:
            raise ValueError("구매 수량은 0보다 큰 정수여야 합니다.")

        total_price = self.price * quantity
        discount_rate = 0.0

        if 11 <= quantity <= 50:
            discount_rate = 0.05
        elif quantity >= 51:
            discount_rate = 0.10

        return total_price * (1 - discount_rate)

    def purchase(self, quantity: int) -> bool:
        """
        제품을 구매하고 재고를 감소시킵니다.

        Args:
            quantity (int): 구매 수량.

        Returns:
            bool: 구매 성공 여부. 재고가 충분하면 True, 부족하면 False.
        """
        if not isinstance(quantity, int) or quantity <= 0:
            raise ValueError("구매 수량은 0보다 큰 정수여야 합니다.")

        if self.stock >= quantity:
            self.stock -= quantity
            print(f"{self.name} {quantity}개 구매 완료.")
            return True
        else:
            print(f"재고가 부족합니다. 현재 재고: {self.stock}")
            return False

# --- 테스트 코드 ---
if __name__ == "__main__":
    # 단가 1000원인 "mask"를 10,000장으로 설정
    mask = Articles(name="mask", stock=10000, price=1000)

    try:
        # 40장 구매 가격 출력
        quantity_to_buy = 40
        total_cost = mask.offer_price(quantity_to_buy)
        print(f"{mask.name} {quantity_to_buy}장 구매 시 총 가격 (5% 할인 적용): {total_cost:,.0f}원")

        # 40장 구매 후 재고 출력
        print(f"구매 전 재고: {mask.stock:,}개")
        mask.purchase(quantity_to_buy)
        print(f"구매 후 재고: {mask.stock:,}개")

    except ValueError as e:
        print(f"오류 발생: {e}")

```
